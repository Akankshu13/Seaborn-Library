```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=[1,5,6,4,3,8,2]
var_1=[6,1,2,3,9,6,4]

x_1=pd.DataFrame({"var":var,"var_1":var_1})

sns.lineplot(x="var",y="var_1",data=x_1)   #data store the value of x_1
```

```python
#if we want to remove (<Axes: xlabel='var', ylabel='var_1'>) ,we use plt.show

var=[1,5,6,4,3,8,2]
var_1=[6,1,2,3,9,6,4]

x_1=pd.DataFrame({"var":var,"var_1":var_1})

sns.lineplot(x="var",y="var_1",data=x_1)   #data store the value of x_1

plt.show()
```

```python
#if we do not want to use dataframe

var=[1,5,6,4,3,8,2]
var_1=[6,1,2,3,9,6,4]

sns.lineplot(x=var,y=var_1)   #here we pass value directly to x and y
plt.show()
```

```python
da=sns.load_dataset("planets")            #here we use github to load data with the help of sns.load_dataset function and in () we pass which data we want
                                          # for github we search on google seaburn data and go to github link
                                          #  planets.csv, this is the data bcz we use directly so we remove csv
                                           # we use da bcz we want to store our data to da
da          (#we want to see the data)       

```

```python
sns.lineplot(x="year",y="distance",data=da)   #here we pass value in x and y which comparison we want to show our graph and in data we write which data we use 

plt.show()
```

```python
sns.lineplot(x="year",y="distance",data=da,hue="method")   #here we pass value in x and y which comparison we want to show our graph and in data we write which data we use 
                                #if we want to make graph in respect of powerful data so we use hue
plt.show()                      #In other words,'hue' is used to visualize the data of different categories in one plot.
```

```python
#if we Want to change size, then we use size parameter

sns.lineplot(x="year",y="distance",data=da,hue="method",size=0.2) 
plt.show()
```

```python
sns.lineplot(x="year",y="distance",data=da,hue="method",style="method")     #style method change style
plt.show()
```

```python
#if we want to change color
#palette' is used to change the colour of the plot.

sns.lineplot(x="year",y="distance",data=da,hue="method",palette="Spectral")     #style method change style
plt.show()
```

```python
sns.lineplot(x="year",y="distance",data=da,hue="method",markers=['o'] )    #style method change style
plt.show()
```

```python
#if we want to work less data

da=sns.load_dataset("planets").head(20) 
sns.lineplot(x="year",y="distance",data=da,hue="method",markers=['+'],style="method" )    #style method change style
plt.show()
```

```python
#if we want to remove dotted line we use dashes function
da=sns.load_dataset("planets").head(20) 
sns.lineplot(x="year",y="distance",data=da,hue="method",markers=['+'],style="method",dashes=False )    #style method change style
plt.show()
```

```python
#if we want to remove legend

da=sns.load_dataset("planets").head(20) 
sns.lineplot(x="year",y="distance",data=da,hue="method",style="method",legend=False )    #style method change style
plt.show()
```

```python
#legend must be auto,brief,full or a boolean

da=sns.load_dataset("planets").head(20) 
sns.lineplot(x="year",y="distance",data=da,hue="method",style="method",legend="brief" )    
```

```python

#if we want grid

da=sns.load_dataset("planets").head(20) 
sns.lineplot(x="year",y="distance",data=da,hue="method",style="method",legend="brief" )
plt.grid()
```

```python
#if we want to add title

da=sns.load_dataset("planets").head(20) 
sns.lineplot(x="year",y="distance",data=da,hue="method",style="method",legend="brief" )
plt.title("Planets")
plt.show()
```

# Another Example

```python
data=sns.load_dataset("healthexp")   
data
```

```python
sns.lineplot(x="Year",y="Spending_USD",data=data,hue="Country",style="Country",markers=False)  
plt.show()
```

# Another Example

```python
a=sns.load_dataset("flights")
a
```

```python
sns.lineplot(x="year",y="passengers",data=a,markers=False,legend=True)  

plt.show()
```

```python
sns.lineplot(x="year",y="passengers",data=a,markers=False,legend=True,hue="month",style="month")  

plt.show()
```

```python
sns.lineplot(x="month",y="passengers",data=a,markers=False,legend=True)
plt.show()
```

```python
sns.lineplot(x="month",y="passengers",data=a,markers=False,legend=True,hue="year")
plt.show()
```

```python
sns.lineplot(x="month",y="passengers",data=a,markers=False,legend=True,hue="year",style="year")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.show()
```
