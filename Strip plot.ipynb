```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df
```

```python
sns.stripplot(x="day",y="total_bill",data=df)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=df,hue="sex")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=df,palette="ocean",hue="sex")
plt.show()
```

```python
#draw the underline of circle with the help of linewidth
sns.stripplot(x="day",y="total_bill",data=df,palette="ocean",hue="sex",linewidth=1)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=df,palette="ocean",hue="sex",edgecolor="r")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=df,palette="ocean",hue="sex",linewidth=1,jitter=8)
plt.show()
```

```python
#increase the size with the help of size parameter
sns.stripplot(x="day",y="total_bill",data=df,palette="ocean",hue="sex",size=10)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=df,palette="ocean",hue="sex",marker=">")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=df,palette="ocean",hue="sex",alpha=0.5)
plt.show()
```

```python
sns.stripplot(x="total_bill",data=df)
plt.show()
```
