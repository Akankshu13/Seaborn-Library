```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=df)
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="time")
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="sex")
```

```python
#if we want chamge color
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="sex",color="b")
```

```python
#we can change order also
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="time",order=["Sat","Sun","Thur","Fri"])
```

```python
#show mean
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="sex",showmeans=True)
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=df,showmeans=True,color="y")
```

```python
#here we change marker
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="sex",showmeans=True,meanprops=({"marker":"+","markeredgecolor":"b"}))
```

```python
#change line width

sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="sex",showmeans=True,linewidth=3)
```

```python
#change color
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="sex",showmeans=True,palette="plasma")
```

```python
#if we want horizontal
sns.set(style="whitegrid")
sns.boxplot(x="total_bill",y="day",data=df,hue="sex",showmeans=True)
```

```python
#if we want single box Plot
sns.set(style="whitegrid")
sns.boxplot(x=df["total_bill"])
```
