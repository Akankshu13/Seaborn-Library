```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#here we create 2D array
df=np.linspace(1,30,20).reshape(4,5)
df
```

```python
#here we create heatmap
sns.heatmap(df)
```

```python
data=sns.load_dataset("anagrams")
data
```

```python
#it take numerical value, so we drop the object column and store its value to some other variable
data=data.drop(columns=["attnr"],axis=1)   #if we want to drop multiple columns then we can also do that
data
```

```python
data
```

```python
sns.heatmap(data)
plt.show()
```

```python
#if we want to make a graph with less data
sns.heatmap(data.head(10))
plt.show()
```

```python
#we can also set min and max value
sns.heatmap(data.head(10),vmin=1,vmax=6)
plt.show()
```

```python
#we can also change the colour
sns.heatmap(data.head(10),cmap="cividis")
plt.show()
```

```python
#here we show the value of graph with the help of annot parameter
sns.heatmap(data.head(10),annot=True)
plt.show()
```

# Another Example

```python
df6=np.linspace(1,10,10).reshape(2,5)
df6
```

```python
sns.heatmap(df6,annot=True)
plt.show()
```

```python
#here we set value with the help of array
arr=np.array([["a0","a1","a2","a3","a4"],
              ["b0","b1","b2","b3","b4"]])
arr
```

```python
sns.heatmap(df6,annot=arr,fmt="s")
plt.show()
```

```python
#we can also change the font size
y={"fontsize":50,"color":"r"}
sns.heatmap(df6,annot=True,annot_kws=y)
plt.show()
```

```python
#we can also change the width of box
sns.heatmap(df6,annot=True,linewidth=4)
plt.show()
```

```python
#we can also change the color of boundary
sns.heatmap(df6,annot=True,linewidth=4,linecolor="b")
plt.show()
```

```python
#we can also remove the colorbar
sns.heatmap(df6,annot=True,cbar=False)
plt.show()
```

```python
#here we remove x label
sns.heatmap(df6,annot=True,xticklabels=False)
plt.show()
```

```python
#we can also remove y label
sns.heatmap(df6,annot=True,yticklabels=False)
plt.show()
```

```python
#if we want to set label
v=sns.heatmap(df6,annot=True)
v.set(xlabel="Python",ylabel="C")
plt.show()
```

```python
v=sns.heatmap(df6,annot=True)
v.set(xlabel="Python",ylabel="C")
sns.set(font_scale=50)
plt.show()
```
