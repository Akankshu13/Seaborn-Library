```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### Difference between count plot and bar plot

```python
df=sns.load_dataset("tips")
df
```

```python
sns.countplot(x="sex",data=df)
plt.show()
```

```python
sns.barplot(x="sex",y="size",data=df)
plt.show()
```

```python
sns.countplot(x="sex",data=df,hue="smoker")
plt.show()
```

```python
#if we want horizontal graph
sns.countplot(y="sex",data=df,hue="smoker")
plt.show()
```

```python
sns.countplot(x="sex",data=df,hue="smoker",palette="vlag")
plt.show()
```

```python
sns.countplot(x="sex",data=df,hue="smoker",palette="vlag",alpha=0.5)
plt.show()
```

```python
sns.countplot(x="sex",data=df,hue="smoker",palette="vlag",saturation=100)
plt.show()
```
