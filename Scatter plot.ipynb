```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("penguins")
df
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df)
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex")
plt.show()
```

```python
#here we change the style using style parameter
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",style="sex")
plt.show()
```

```python
#here we changes the size
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",size="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",))
plt.show()
```

```python
#here we change colour
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",palette="gist_ncar")
plt.show()
```

```python
#here we blur the color using alpha parameter
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",alpha=0.5)
plt.show()
```
