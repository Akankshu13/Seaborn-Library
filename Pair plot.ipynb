```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dff=sns.load_dataset("tips")
dff
```

```python
sns.pairplot(dff)
```

```python
#if we want to differentiate according to sex
sns.pairplot(dff,hue="sex")
```

```python
#if we want only two column in graph
sns.pairplot(dff,hue="sex",vars=["total_bill","tip"])
```

```python
#we can also set the order of hue
sns.pairplot(dff,hue="sex",vars=["total_bill","tip"],hue_order=["Female","Male"])
```

```python
#colour change
sns.pairplot(dff,hue="sex",vars=["total_bill","tip"],palette="spring")
```

```python
#if we wantto change graph
# we can put kde,hist,scattter,reg in kind
sns.pairplot(dff,hue="sex",vars=["total_bill","tip"],kind="reg")
```

```python
#if we want to change dia plot shape
#we can only put kde or hist in diag kind
sns.pairplot(dff,hue="sex",vars=["total_bill","tip"],diag_kind="hist")
```

```python
#if we want to change ,ale female markers
sns.pairplot(dff,hue="sex",vars=["total_bill","tip"],markers=[">","*"])
```
