```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a=sns.load_dataset("penguins")
a
```

```python
sns.barplot(x="island",y="bill_length_mm",data=a)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=a,hue="sex")
plt.show()
```

```python
orders=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=a,hue="sex",order=orders)
plt.show()
```

### if we want feamale first in graph then we can also do this with the help of hue order parameter

```python
sns.barplot(x="island",y="bill_length_mm",data=a,hue="sex",hue_order=["Female","Male"])
plt.show()
```

### we can also set percentile line using ci parameter

```python
sns.barplot(x="island",y="bill_length_mm",data=a,hue="sex",hue_order=["Female","Male"],ci=10)
plt.show()
```

### with the help of orient parameter we can draw vertical graph but always give numeric value

```python
sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=a,orient="h")
plt.show()
```

### if we want to change color

```python
sns.barplot(x="island",y="bill_length_mm",data=a,hue="sex",hue_order=["Female","Male"],palette="viridis")
plt.show()
```

### if we want to do color blur then we will use saturation

```python
# sns.barplot(x="island",y="bill_length_mm",data=a,hue="sex",hue_order=["Female","Male"],palette="viridis",saturation=0.2)
plt.show()
```

### change error color

```python
sns.barplot(x="island",y="bill_length_mm",data=a,hue="sex",hue_order=["Female","Male"],palette="viridis",errcolor="g")
plt.show()
```

### Increase error width

```python
sns.barplot(x="island",y="bill_length_mm",data=a,hue="sex",hue_order=["Female","Male"],palette="viridis",errcolor="g",errwidth=5)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=a,hue="sex",hue_order=["Female","Male"],palette="viridis",errcolor="g",capsize=0.05)
plt.show()
```

### with the help of alpha parameter we can light the color

```python
sns.barplot(x="island",y="bill_length_mm",data=a,hue="sex",hue_order=["Female","Male"],palette="viridis",errcolor="g",alpha=0.5)
plt.show()
```

### we can also make grid

```python
sns.set(style="darkgrid")
sns.barplot(x="island",y="bill_length_mm",data=a,hue="sex",hue_order=["Female","Male"],palette="viridis",errcolor="g")
plt.show()
```
